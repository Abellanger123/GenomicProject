```r
#load and format the clinical data

clinical_data <- read.table("clinical_data.tsv", header = TRUE, sep = "\t")
clinical_data$COHORT <- as.factor(clinical_data$COHORT)
clinical_data$DTHHRDY <- as.factor(clinical_data$DTHHRDY)
clinical_data$SEX <- as.factor(clinical_data$SEX)
clinical_data
```

```r
#load and format the morphological counts data

morpho_counts <- read.table("morphological_counts_lunit_dino.tsv", 
                           sep = "\t", header=TRUE)

row.names(morpho_counts) <- morpho_counts$SMPLID
morpho_counts <- morpho_counts[-1]
morpho_counts
```

```r
#transpose the morpholigical counts (samples as columns, clusters as rows)

morpho_counts2 <- t(morpho_counts)
morpho_counts2
ncol(morpho_counts2)
```

Simple analysis : sex

```r
#age binary variable : under or equal to 55, above 55
age_fact <- ifelse(clinical_data$AGE < 56, 1, 2)
age_fact
```

```r
#height binary variable : under or equal to 68, above 68
median(clinical_data$HGHT)
hght_fact <- ifelse(clinical_data$HGHT < 68, 1, 2)
hght_fact
```

```r
#weight binary variable : under or equal to 176, above 176
median(clinical_data$WGHT)
wght_fact <- ifelse(clinical_data$WGHT < 176, 1, 2)
wght_fact
```

```r
#bmi binary variable : under or equal to 27, above 27
median(clinical_data$BMI)
bmi_fact <- ifelse(clinical_data$BMI < 27, 1, 2)
bmi_fact
```

```r
#TRISCHD binary variable : under or equal to 442, above 442
median(clinical_data$TRISCHD)
trischd_fact <- ifelse(clinical_data$TRISCHD < 442, 1, 2)
trischd_fact
```

```r
metaData <- data.frame(COHORT=as.numeric(clinical_data$COHORT), SEX=as.numeric(clinical_data$SEX),
                      DTHHRDY=as.numeric(clinical_data$DTHHRDY), AGE_FACT=age_fact, HGHT_FACT=hght_fact,
                      WGHT_FACT=wght_fact, BMI_FACT=bmi_fact, TRISCHD_FACT=trischd_fact)
row.names(metaData) <- clinical_data$SMPLID
metaData
```

```r
library(DESeq2)
```

```r
ncol(morpho_counts2)
nrow(metaData)
metaData
```

```r
dds_sex <- DESeqDataSetFromMatrix(countData = round(morpho_counts2),
                              colData = round(metaData),
                              design = ~ SEX)
dds_cohort <- DESeqDataSetFromMatrix(countData = round(morpho_counts2),
                              colData = round(metaData),
                              design = ~ COHORT)
dds_dthhrdy <- DESeqDataSetFromMatrix(countData = round(morpho_counts2),
                              colData = round(metaData),
                              design = ~ DTHHRDY)
dds_age_fact <- DESeqDataSetFromMatrix(countData = round(morpho_counts2),
                              colData = round(metaData),
                              design = ~ AGE_FACT)
dds_hght_fact <- DESeqDataSetFromMatrix(countData = round(morpho_counts2),
                              colData = round(metaData),
                              design = ~ HGHT_FACT)
dds_wght_fact <- DESeqDataSetFromMatrix(countData = round(morpho_counts2),
                              colData = round(metaData),
                              design = ~ WGHT_FACT)
dds_bmi_fact <- DESeqDataSetFromMatrix(countData = round(morpho_counts2),
                              colData = round(metaData),
                              design = ~ BMI_FACT)
dds_trischd_fact <- DESeqDataSetFromMatrix(countData = round(morpho_counts2),
                              colData = round(metaData),
                              design = ~ TRISCHD_FACT)
```

```r
dds_sex
dds_cohort
dds_dthhrdy
dds_age_fact
dds_hght_fact
dds_wght_fact
dds_bmi_fact
dds_trischd_fact
```

```r
dds_sex <- DESeq(dds_sex)
dds_cohort <- DESeq(dds_cohort)
dds_dthhrdy <- DESeq(dds_dthhrdy)
dds_age_fact <- DESeq(dds_age_fact)
dds_hght_fact <- DESeq(dds_hght_fact)
dds_wght_fact <- DESeq(dds_wght_fact)
dds_bmi_fact <- DESeq(dds_bmi_fact)
dds_trischd_fact <- DESeq(dds_trischd_fact)
```

```r
#clinical variable
#clusters 3 and 4 associated with sex
res_sex <- results(dds_sex)
res_sex <- res_sex[order(res_sex$padj),]
res_sex <- data.frame(res_sex)
res_sex
```
